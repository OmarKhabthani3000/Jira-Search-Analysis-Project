```python
import requests
from requests.packages.urllib3.exceptions import InsecureRequestWarning
# Base URL for your Jira instance
base_url = "https://hibernate.atlassian.net/rest/api/3"
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
# Authentication credentials
username = "omar.khabthani123@gmail.com"
api_token = "ATATT3xFfGF0Z-pgdOKTQWqsYiGVkZRzYkGKQVoAF0yg_mRWo4qXWjTZY6monU7oST4v9IpDHSPfahsokbglvpD6EFxNpMX2bA0P2NFORrfwxqsTrfsj6NIJHThqlqZ47K61VN6P4LljFTS07Egfw-BdKdXHHv-8uI2K2VRS_pql8D2jPOJU2iY=6224B1E1"
auth = (username, api_token)

# Function to generate issue URL
def get_issue_url(issue_key):
    return f"{base_url}/issue/{issue_key}"

# Function to generate comments URL
def get_comments_url(issue_key):
    return f"{base_url}/issue/{issue_key}/comment"

# Function to generate changelog URL
def get_changelog_url(issue_key):
    return f"{base_url}/issue/{issue_key}/changelog"

# Function to generate attachments URL
def get_attachments_url(issue_key):
    return f"{base_url}/issue/{issue_key}/attachments"

# Jira API endpoint for searching issues
search_url = f"{base_url}/search"

# Project key to filter issues
project_key = "BVAL"

# JQL query to retrieve all issues in the specified project
jql_query = f"project = '{project_key}'"

# Parameters for the API request
params = {
    "jql": jql_query,
    "maxResults": 1000  # Adjust as needed based on the number of issues
}
headers = {
    "Accept": "application/json",
    "Content-Type": "application/json"
}
# Make the request
response = requests.get(search_url, headers=headers, params=params, auth=auth, verify=False)

# Check if the request was successful
if response.status_code == 200:
    # Parse JSON response
    data = response.json()
    
    # Loop through all issues
    for issue in data["issues"]:
        issue_key = issue["key"]
        issue_summary = issue["fields"]["summary"]
        
        # Get issue details
        issue_details_response = requests.get(get_issue_url(issue_key), auth=auth, verify=False)
        issue_details = issue_details_response.json()
        
        # Get issue comments
        comments_response = requests.get(get_comments_url(issue_key), auth=auth, verify=False)
        comments = comments_response.json()["comments"]
        
        # Get issue changelog (history)
        changelog_response = requests.get(get_changelog_url(issue_key), auth=auth, verify=False)
        changelog = changelog_response.json()["values"]
        
        
        # Process the data as needed
        # For example, print issue details, comments, history, etc.
        print("Issue Key:", issue_key)
        print("Issue Summary:", issue_summary)
        #print("Issue Details:", issue_details)
        print("Issue Comments:", comments)
        print("Issue Changelog:", changelog)
else:
    print("Error:", response.status_code, response.text)
```
